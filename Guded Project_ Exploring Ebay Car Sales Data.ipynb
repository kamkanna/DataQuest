## About the project
The aim of this priject is to clean the data and analyze the included used car listings.
The used data set from Ebay Kleinzeigen, a classifyied section of German Website. The data set was originally scraped and uploaded to Kaggle.

The data dictionary provided with data is as follows:

* dateCrawled - When this ad was first crawled. All field-values are taken from this date.
* name - Name of the car.
* seller - Whether the seller is private or a dealer.
* offerType - The type of listing
* price - The price on the ad to sell the car.
* abtest - Whether the listing is included in an A/B test.
* vehicleType - The vehicle Type.
* yearOfRegistration - The year in which year the car was first registered.
* gearbox - The transmission type.
* powerPS - The power of the car in PS.
* model - The car model name.
* kilometer - How many kilometers the car has driven.
* monthOfRegistration - The month in which which year the car was first registered.
* fuelType - What type of fuel the car uses.
* brand - The brand of the car.
* notRepairedDamage - If the car has a damage which is not yet repaired.
* dateCreated - The date on which the eBay listing was created.
* nrOfPictures - The number of pictures in the ad.
* postalCode - The postal code for the location of the vehicle.
* lastSeenOnline - When the crawler saw this ad last online.

## Investigation of registration year 

The outliers of registration column should be removed. Before it will br done let's seew what's the percentage of the data has odd values.
The first car was invented in 1895. Therefore, the year range should be 1895 - 2016.

```python
(~autos["registration_year"].between(1900,2016)).sum() / autos.shape[0]
```

There are 4% listings out of the year range. Therefore, these listings will be removed.

```python
autos = autos[autos["registration_year"].between(1895,2016)]
print(autos["registration_year"].describe())
autos["registration_year"].value_counts(normalize=True)
```

The more cars of this distribution have been registered in the period 1994-2016.

## Exploring Price by Brand

```python
print(autos["brand"].unique())
autos["brand"].value_counts(normalize=True)
```

Despite the fact that half of listings presented by German brends. Let's look at the prices of top-10 brands. 
- volkswagen
- bmw
- opel
- mercedes_benz
- audi
- ford
- renault
- peugeot
- fiat
- seat

```python
brand = autos["brand"].value_counts().head(10).index
brand_mean_price = {}

for b in brand:
    selected_rows = autos[autos['brand'] == b]
    mean_price = selected_rows['price'].mean()
    brand_mean_price[b] = int(mean_price)
    
brand_mean_price
```

There is a price gap between popular brend:
- The last top-5 brends, Seat, Feat, Peugeot, Renault, Ford are in price rnge of \$2474 - 4397. 
- The most expensive brands are Audi, BMW, Mercedes-Benz with average price of \$8300-9300. 
- The most popular brens, Volkswagen, has average price of \$5400. That is between the cheap brands and expensive ones. It may be the best option of cheap and expensive worlds.

## Exploring Mileage

For the top 10 brands, let's use aggregation to understand the average mileage for those cars and if there's any visible link with mean price.

```python
bmp_series = pd.Series(brand_mean_price)
bmp_series
```

```python
pd.DataFrame(bmp_series, columns=['mean_mileage'])
```

```python
brand_mean_mileage = {}

for b in brand:
    selected_rows = autos[autos['brand'] == b]
    mean_mileage = selected_rows['odometer_km'].mean()
    brand_mean_mileage[b] = int(mean_mileage)
    
brand_mean_mileage
```

```python
mean_price = pd.Series(brand_mean_price).sort_values(ascending=False)
mean_mileage = pd.Series(brand_mean_mileage).sort_values(ascending=False)
```

```python
brand_mean_price_mileage = pd.DataFrame(mean_mileage, columns=['mean_mileage'])
brand_mean_price_mileage
```

```python
brand_mean_price_mileage['mean_price'] = mean_price
brand_mean_price_mileage
```

There is not a big gap in mileage of the top brands. Theere is a mileage trend. The more expensive car the higher th mileage.
